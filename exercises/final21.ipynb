```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def c(x, M):
    y = np.ones(len(x))
    for n in range(1, M+1):
        y = y + 2 * np.cos(n * x)
    return y
```

```python
def s(x, M):
    return np.sin(x * (2*M + 1) / 2.0) / np.sin(x / 2.0)
```

```python
M=5
plt.plot(c(np.arange(0, 6, .01), M))
plt.plot(s(np.arange(0, 6, .01), M))
```

```python
def err(sigma):
    M = np.ceil(9 * sigma * sigma / 4)
    M = int(np.floor(M/2) * 2)
    print(M)
    TRIALS = 100000
    ko = 0
    a = np.ones(M)
    b = np.array([1 if n % 2 == 0 else -1 for n in range(0, M)])
    for n in range (0, TRIALS):
        eta = sigma * np.random.randn(M)
        ta = np.sum((a + eta) * a)
        tb = np.sum((a + eta) * b)
        #print(ta, tb)
        if  ta < tb:
            ko += 1
    print(ko / TRIALS)
```

```python
err(100)
```

```python
def sincint(N, delta):
    t = N * (1 / (np.pi * delta) + 1)
    for tt in np.arange(t, 0, -.1):
        x = 0
        for n in range(0, N):
            x = x + np.sin(np.pi * (tt - n)) / (np.pi * (tt - n))
        #print(x)
        if np.abs(x) > delta:
            break
    print(t, tt)
```

```python
sincint(10, 0.001)
```
