Simple iid process, +1 or -1 with 50% probability. Clearly ergodic.

Create N realization with N samples each and arrange as a matrix. Plot matrix as image. If ergodic, plotting the transposed matrix should give a visually equivalent image. This seems to be the case

```python
N=100
n = np.arange(0,N)
y = np.ones(N)
x = np.zeros((N, N))
for i in xrange(0, N):
    x[i] = np.sign(np.random.randn(1, N))
plt.subplot(1,2,1).imshow(x)
plt.subplot(1,2,2).imshow(x.T)
```

Now let's filter each realization process with a simple lowpass filter. If filter preserves ergodicity, normal and transposed image should look "equivalent" but this is not the case. Sample size too small?

```python
for i in xrange(0, N):
    x[i] = sp.lfilter([1, 1, 1], 1, x[i])
plt.subplot(1,2,1).imshow(x)
plt.subplot(1,2,2).imshow(x.T)
```

these are the per-row averages on normal and transposed matrices; again, qualitatively different

```python
plt.plot(sum(x, 1)/N)
```
